1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

Answer -  Covariance is a statistical measure that indicates the direction of the linear relationship between two random variables.

Difference between covariance anf correlation-

Scale - Covariance depends on the units of variables and It can range from -∞ to +∞. Correlation is dimensionless value and it is always "normalized" to range between -1 to +1.

Interpretation - Covariance only tells us the direction of the relationship while correlation tells direction and the strength of the relationship.

2. What does a positive, negative, and zero covariance indicate about the relationship between two variables.

Answer -Covaraince is a measure of how two random variables change together. The sign of the covariance indicates the direction of the linear relationship between the variables.

* Positive Covariance - Indicates that the two variables tend to move in the same direction. As the value of one variable increases, the value of the other variable also tends to increase.

* Negative Covariance- Indicates that the two variables tend to move in opposite directions. As the value of one variable increases, the value of the other variable tends to decrease.

* Zero Variance - Indicates that there is no linear relationship between two variables.

3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

Answer - Limitations Of Coveriance -

* Scale dependency - The value of covariance is highly sensitive to the units of measurement. If we change the units of our data from meters to centimeters the convariance value will change significantly, even though the underlying relationship remains the same.

* Lack of standardised range - Covariance can range from negative infinity to positive infinity. Because there is no fixed upper or lower bound.

*  Difficulty in comparison - We cannot easily compare the covariance of two different pairs of variables if they are measured in different units.

Reason behind also preferred correlation -

* Covariance always provides a range between -1 to +1 . The fixed scale allows us to instantly judge the strength of the connection.

* It is standardized, we can compare the strength of relationships across entirely different datasets.

4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?

Answer - difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficien -


* Relationship Type - Pearson's correlation measures linear relationships  while spearman's rank correlation measures monotonic relationships.

* Data Type - Pearson's correation best for interval or ratio data while spearman's rank correlation best for ordinal data.

* Sensitivity to outliers - Pearson's correlation is highly sensitive - a single outlier can significantly change the result. Spearman's rank correlation
is robust - it uses ranks instead of actual values, outliers have less impact.

* Assumptions - Pearson's correlation assumes data is normally distributed and has a linear relationship. Spearman's rank correlation is distribution free does not assume normality.

When we can prefer to use Spearman’s correlation -

* When the relationship between variables is monotonic but not a straight line.

* When our data consists of rankings or scales rather than precise measurements.

* If our dataset contains extreme values that would skew a linear average, spearman is a safer choice because it only cares about the relative order of the data points.

* When our data doesn't follow a bell curve and violates the assumpations required for pearson.
